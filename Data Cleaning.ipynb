```python
# Some of the columns have spaces at the end. I am going to remove them
df.rename(columns=lambda x:x.strip(),inplace=True)
```

```python
#Checking the column names again
df.columns
```

```python
# Next I will be creating a function which will count the null and zero values in every column in the dataset
def countofnulland0(data):
    for column in data.columns.to_list():
        print(column)
        print("Null: ",data[column].isnull().sum())
        print("zeros: ",(data[column]==0).sum(axis=0))
        print('\n')
countofnulland0(df)
```

```python
# Above we can see that there are several null and zero values. We need to decide which zero value to keep.
#For eg: infant deaths column has 848 zero values. So, we need to keep them
```

```python
#Columns whose 0 values will be replaced by the mean of that column
colname=['percentage expenditure','Income composition of resources']
for col in colname:
    print(col)
    print("Original Count of zero values")
    print("Zero Values: ", (df[col]==0).sum())
    print('\n')
    df[col] = np.where(df[col]==0, df[col].mean(), df[col])
    df[col] = df[col].fillna(df[col].mean())
    print("Cleaned Count")
    print("Zero Values: ", (df[col]==0).sum())
    print('\n')
```

```python
#Replacing Null Values with the mean of that column
for column in list(df.select_dtypes(include='float64' or 'int64').columns):
    print(column)
    print("Original Count")
    print("Null Values: ", df[column].isnull().sum())
    print('\n')
    df[column] = df[column].fillna(df[column].mean())
    print("Cleaned Count")
    print("Null Values: ", df[column].isnull().sum())
    print('\n')
```

```python
# Now all the null values and zero values have been handled. Let us check again
df.isna().sum()
```
